**FILLING THE NULL VALUES**

```python
import pandas as pd
```

```python
df.fillna(method = 'ffill')   # it is forward fill means that it will fill your null values with your previous value (comparing to rows , because by default axis = 0)
```

```python
df.fillna(method = 'ffill' , axis =1)   # it is forward fill means that it will fill your null values with your previous value (comparing to columns , because axis = 1)
```

```python
df.fillna(method = 'bfill')   # it is backward fill means that it will fill your null values with your next value
```

```python
df['Physics'].fillna(value=df['Physics'].mean())   # fill the null values with the mean of physics marks in the physics column
```

**merge() FUNCTION --> similar to join of SQL**

```python
df1 = pd.DataFrame({'Roll No.':[1,2,3,4,5] , 'Physics':[20,30,40,50,89]})
df1
```

```python
df2 = pd.DataFrame({'Roll No.':[1,2,3,4,5] , 'Chemistry':[78,33,39,81,90]})
df2
```

```python
pd.merge(df1,df2,on = 'Roll No.')      # default join
```

```python
pd.merge(df2,df1, on = 'Roll No.')
```

```python
pd.merge(df1,df2)     # on --> by default intersecting column is taken
```

```python
df3 = pd.DataFrame({'Roll No.':[1,2,3,4,5] , 'Physics':[20,30,40,50,89]})
df4 = pd.DataFrame({'Roll No.':[1,2,3,9,10] , 'Chemistry':[78,33,39,81,90]})
```

```python
pd.merge(df3,df4)
```

```python
pd.merge(df3,df4, how = 'left')         # left join
```

```python
pd.merge(df3,df4, how = 'right')         # right join
```

```python
pd.merge(df3,df4, how = 'outer')         # outer join
```

**concat() FUNCTION --> similar to UNION of SQL**

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'Roll No.':[1,2,3,4,5] , 'Physics':[20,30,40,50,89] , 'Maths':[45,78,45,90,56]})
df2 = pd.DataFrame({'Roll No.':[6,7,8,9,10] , 'Physics':[28,35,44,52,80] , 'Maths':[45,78,58,57,86]})
print(df1)
print(df2)
```

```python
pd.concat([df1,df2])       # previously append() is used but nowadays concat is used
```

```python
pd.concat([df2,df1])
```

```python
pd.concat([df1,df2] , ignore_index = True)       # ignore_index=True  resets the index in the combined DataFrame.
```

```python
pd.concat([df1,df2] , ignore_index = True ,sort = True)      # sort is going to sort my column names
```

```python
df1 = pd.DataFrame({'Roll No.':[1,2,3,4,5] , 'Chemistry':[20,30,40,50,89] , 'Maths':[45,78,45,90,56]})
df2 = pd.DataFrame({'Roll No.':[6,7,8,9,10] , 'Physics':[28,35,44,52,80] , 'Maths':[45,78,58,57,86]})
print(df1)
print(df2)
```

```python
pd.concat([df1,df2] , ignore_index = True , sort = True)
```

**PIVOT TABLE**

```python
import pandas as pd
```

```python
df = pd.read_csv(r'/content/sample2.csv')
df.head()
```

```python
pd.pivot_table(df, index = 'Branch' , values = ['Physics','Chemistry','Maths','Computer'])    # by default it will give mean values
```

```python
pd.pivot_table(df, index='Branch' , aggfunc='sum', values = ['Physics','Chemistry','Maths','Computer'])
```

```python
pd.pivot_table(df, index='Branch' , aggfunc='count', values = ['Physics','Chemistry','Maths','Computer'])
```

```python
pd.pivot_table(df, index='Branch' , aggfunc='max', values = ['Physics','Chemistry','Maths','Computer'])
```

```python
pd.pivot_table(df, index='Branch' ,columns = 'Section', values = ['Physics','Chemistry','Maths','Computer'])     # mean values in the table (by default)
```
