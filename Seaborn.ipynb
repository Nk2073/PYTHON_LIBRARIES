**BAR PLOT**

```python
titanic_df.head()
```

```python
sns.barplot(x='class' , y ='fare', data = titanic_df)
```

A Bar plot represents an estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bars.

```python
sns.barplot(x='class' , y ='fare', data = titanic_df ,
            hue = 'sex')
```

```python
sns.barplot(x='class' , y ='fare', data = titanic_df ,
            hue = 'sex',
            palette = 'magma')
```

```python
#orient :- orient controls the orientation of your plot — whether the data is plotted vertically or horizontally.
sns.barplot(x='class' , y ='fare', data = titanic_df ,
            hue = 'sex',
            palette = 'magma',
            orient = 'h')
```

```python
#estimator :- estimator is the function used to summarize your data within each category before plotting. By default it is mean.
sns.barplot(x='class' , y ='fare', data = titanic_df ,
            hue = 'sex',
            palette = 'magma',
            estimator = np.max)
```

```python
#confidence interval(ci) :- It is a range of values that’s likely to contain the true value of some estimate (like a mean) with a certain level of confidence — often 95% by default.
sns.barplot(x='class' , y ='fare', data = titanic_df ,
            hue = 'sex',
            palette = 'magma',
            ci = 100,
            errcolor= 'red',
            errwidth = 5)
```

```python
#saturation
sns.barplot(x='class' , y ='fare', data = titanic_df ,
            hue = 'sex',
            palette = 'magma',
            saturation = 0.3    # 30% colorful
            )
```

**HEATMAPS**

```python
flight_df = sns.load_dataset('flights')
flight_df.head()
```

```python
flight_df = flight_df.pivot(index="month", columns="year", values="passengers")
flight_df.head()
```

```python
plt.figure(figsize = (12,6))
ax = sns.heatmap(flight_df)
ax
```

```python
plt.figure(figsize = (12,6))
ax = sns.heatmap(flight_df ,
                 annot=True,    # annotations
                 fmt = 'd'       # controls the format of annotations(integers(d) , float(.1f), general(g))
                 )
ax
```

```python
plt.figure(figsize = (12,6))
ax = sns.heatmap(flight_df , annot=True ,
                 fmt = 'd',
                 linecolor = 'k',     # k means black color
                 linewidth = '5'      # this is the width of the gap
                 )
ax
```

```python
plt.figure(figsize = (12,6))
ax = sns.heatmap(flight_df , annot=True ,
                 fmt = 'd',
                 linecolor = 'k',
                 linewidth = '5',
                 cmap = 'Blues'     # cmap is colormap, it is just a set of colors
                 )
ax
```

```python
plt.figure(figsize = (12,6))
ax = sns.heatmap(flight_df,
                 cbar = False     # cbar = color bar → the little side legend that shows the color scale and what numerical values the colors represent.
                 )
ax
```

```python
grid_kws = {"height_ratios": (.4, .05), "hspace": .4}
f,(ax, cbar_ax) = plt.subplots(2, gridspec_kw = grid_kws)
ax = sns.heatmap(flight_df, cbar_kws={"orientation": "horizontal"}, ax=ax,
                 cbar_ax=cbar_ax)
```
